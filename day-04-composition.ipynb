```python
import numpy as np
# A: shear right (slide top edge rightward)
# B: rotate 90° counter-clockwise

A = np.array([[1, 1],
              [0, 1]])

B = np.array([[0, -1],
              [1, 0]])

v = np.array([1, 0])
```

```python
v_after_B = B @ v
print("After B(rotate):", v_after_B)

v_after_A_after_B = A @ v_after_B
print("After A applied to that: ", v_after_A_after_B)
```

```python
AB = A @ B
print("Composed matrix AB:")
print(AB)

v_via_combined = AB @ v
print("v Transformed by AB in one step:", v_via_combined)
```

```python
BA = B @ A
print("Composed matrix BA:")
print(BA)

v_via_BA = BA @ v
print("v transformed by BA in one step:", v_via_BA)

print("\nIs AB equal to BA?", np.array_equal(AB, BA))
```
