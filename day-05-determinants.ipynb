```python
import numpy as np

A = np.array([[3, 0], [0, 2])
print(f"Predicted area scaling: 3 * 2 = 6")
print(f"NumPy det: {np.linalg.det(A)}")
```

```python
# Cell 2 — Flip across x-axis
B = np.array([[1, 0], [0, -1]])
print(f"Predicted: same magnitude, flipped orientation -> -1")
print(f"NumPy det: {np.linalg.det(B)}")
```

```python
# Cell 3 — Linearly dependent rows (collapse)
C = np.array([[1, 2], [2, 4]])
print(f"Predicted: row 2 = 2 * row 1, space collapses -> 0")
print(f"NumPy det: {np.linalg.det(C)}")
```
