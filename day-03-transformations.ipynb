```python
import numpy as np
A = np.array([[2, 0],
             [0, 2]])

B = np.array([[0, -1],
              [1,  0]])

C = np.array([[1, 1],
              [0, 1]])

D = np.array([[3, 1],
              [1, 2]])

i_hat = np.array([1, 0])
j_hat = np.array([0, 1])

for name, M in [("A", A), ("B", B), ("C", C), ("D", D)]:
    print(f"{name} sends î to {M @ i_hat}, ĵ to {M @ j_hat}")
```
